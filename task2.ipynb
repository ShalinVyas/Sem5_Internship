Create a dataset

```python
data={'Name':['Jay','Ajay','Abhi','Ashish','Jenil','Ashil','Lay','Vidhi','Yash','Darshi'],
     'Age':[20,40,35,15,23,56,89,45,12,63],
     'City':['New York','Paris','London','New York','Paris','London','New York','Paris','London','Seoul'],
      'Salary':[100,200,450,895,156,498,745,489,315,636]}
```

```python
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
avg=df.groupby(['Name']).Salary.mean()
```
